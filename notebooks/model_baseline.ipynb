```python
import pandas as pd
import sys
sys.path.append('../src')
from model_evaluation import split_data, evaluate_model

df = pd.read_csv("../data/processed/train_clean.csv")

X_train, X_val, X_test, y_train, y_val, y_test = split_data(df)
```

```python
from sklearn.linear_model import LogisticRegression

logreg_model = LogisticRegression(max_iter=1000)

logreg_model.fit(X_train, y_train)

evaluate_model(logreg_model, X_val, y_val);
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier

gb_model = HistGradientBoostingClassifier(
    class_weight='balanced',
    max_iter=200,            # 200–500 is common, will stop earlier if using early_stopping
    random_state=123)

gb_model.fit(X_train, y_train);

evaluate_model(gb_model, X_val, y_val);
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(class_weight='balanced', n_estimators=200, max_depth=10, random_state=123)
rf_model.fit(X_train, y_train);

evaluate_model(rf_model, X_val, y_val);
```
